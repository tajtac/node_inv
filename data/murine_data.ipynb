```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
import scipy.io
f = scipy.io.loadmat('PU1_034_3/NEWLYFIXED_PU1_034_3.mat')
```

```python
"""
How to retrieve the data from the matlab file:

f[str1][str2][0,0][0,i][j]

where   str1  ['data_dic_save', 'current_save', 'reference_save']
        str2  ['displacements', 'dispinfo', 'strains', 'straininfo'] 
        i  [0,49]      <- time steps (when str1 = 'data_dic_save' and str2 = 'displacements' or 'strains')
        j  [0,7]       <- Variable to retrieve. [0,...,7] ~ [exx_ref, exy_ref, eyy_ref, roi_ref, 
                                                              exx_cur, exy_cur, eyy_cur, roi_cur] 
                                                              (when str1 = 'data_dic_save' and str2 = 'strains')
"""
i1 = 57
i2 = 236
i3 = 55
i4 = 218
nx = i4-i3
ny = i2-i1

x = np.linspace(0,1,nx)
y = np.linspace(0,1,ny)
x, y = np.meshgrid(x,y)

epsx = np.zeros([50,ny,nx])
epsy = np.zeros([50,ny,nx])
aux1 = np.zeros([50,291,275])
aux2 = np.zeros([50,291,275])
for i in range(50):
    aux1[i] = f['data_dic_save']['strains'][0,0][0,i][0]
    aux2[i] = f['data_dic_save']['strains'][0,0][0,i][2]
    epsx[i] = aux1[i][i1:i2, i3:i4]
    epsy[i] = aux2[i][i1:i2, i3:i4]

t1 = 1
fig, ax = plt.subplots(2,2,figsize=(10,10))
ax[0,0].contourf(aux1[t1], cmap='YlOrRd')
ax[0,1].contourf(x,y,epsx[t1], cmap='YlOrRd')
ax[0,0].set(title='$\epsilon_{xx}$')
ax[0,1].set(title='$\epsilon_{xx}$')


t2 = 49
ax[1,0].contourf(aux2[t2], cmap='YlOrRd')
ax[1,1].contourf(x,y,epsy[t2], cmap='YlOrRd')
ax[1,0].set(title='$\epsilon_{yy}$')
ax[1,1].set(title='$\epsilon_{yy}$')
```

```python
print(np.min(epsx), np.max(epsx))
```
